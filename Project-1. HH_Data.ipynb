# Преобразование данных

1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def eduction_fun(string):
    """Функция-преобразование, которая извлекает из признака "Образование и ВУЗ" категорию образование соискателя

    Args:
        string (string): значение признака

    Returns:
        string: одна из четырех возможных категорий: "высшее", "неоконченное высшее", "среднее специальное" 
                и "среднее", либо неизменный аргумент на случай, если входная строка по какой-то приине не 
                соответствует данным в задании гарантиям 
    """    
    first_split = string.split(' ')[1].lower() #разбираем строку на слова и приводим все к нижнему регистру
    if (first_split == 'высшее') or (first_split == 'специальное'):
        return string.split()[0].lower() + ' ' + first_split
    elif (first_split == 'образование'):
        return string.split()[0].lower()
    return string # страховка от не соответствия гарантиям
            
#добавление нового и удаление старого признака, а также проверка на уникальность
hh_database['Образование'] = hh_database['Образование и ВУЗ'].apply(eduction_fun)
hh_database = hh_database.drop(columns=['Образование и ВУЗ'], axis=1)
hh_database['Образование'].unique()
```

2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
def gender_fun(string):
    """Функция, которая возвращает первую букву в первой подстроке, отделенной от остальных запятыми. 
    Для срок из базы это всегда только "М" или "Ж", поскольку все строки начинаются со слов "Мужчина" или "Женщина"

    Args:
        string (string): значение признака

    Returns:
        string: буква "М" или "Ж"
    """    
    split_string = string.split(' , ')
    return split_string[0][0]

def age_fun(string):
    """Функция, которая вытаскивает из строки возраст. В ее основе лежит разбиение по запятым,
    после чего вторая подстрока разбивается по пробелам. Задание гарантирует, что первый элемент такого 
    разбиения - это цифры возраста 

    Args:
        string (string): значение признака

    Returns:
        int: двух или трехзначное натуральное число
    """    
    split_string = string.split(' , ')
    return int(split_string[1].split(' ')[1])

# формируем новые признаки с помощью написанных функций и удаляем старый
hh_database['Пол'] = hh_database['Пол, возраст'].apply(gender_fun)
hh_database['Возраст'] = hh_database['Пол, возраст'].apply(age_fun)
hh_database = hh_database.drop(columns=['Пол, возраст'], axis=1)
hh_database.info()
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def exp_mon_fun(string):
    """Функция, конвертирующая строку с опытом работы в целое число месяцев, либо NaN
    
    Args:
        string (string): значение признака

    Returns: 
        None: если "Не указано" или None
        int: в прочих случаях
    """    
    
    #сначала отсеиваем возврат None
    if (string == "Не указано") or (pd.isna(string)):
        return None
    str_split = string.split(' ')
    
    #во вложенных условиях определеяем одну из трех ситуаций задания опыта
    if str_split[3][0] == 'м': # если опыт меньше года и исчисляется только месяцами на букву "м", после чего переводим все в месяцы
        months = int(str_split[2])
    else:
        if str_split[5][0:5] == 'месяц': #определяем, полные или неполные года опыта
            months = 12*int(str_split[2]) + int(str_split[4])
        else:
            months = 12*int(str_split[2])
    return months

# формируем новый признак с помощью написанной функции и удаляем старый
hh_database['Опыт работы (месяц)'] = hh_database['Опыт работы'].apply(exp_mon_fun)
hh_database = hh_database.drop(columns=['Опыт работы'], axis=1)
display(hh_database['Опыт работы (месяц)'].median())

hh_database.info()
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Список городов-миллионников:

    * … , готов к командировкам , …

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#для функции ниже готовим списки-категории с городами
msk_spb = ['Москва', 'Санкт-Петербург']
million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]

def city_fun(string):
    """Функция для категоризации места жительства соискателя. Основана на том, что имя города -- это первое слово в строке
    
    Args:
        string (string): значение признака

    Returns:
        string: одна из четырех категорий: "Москва", "Санкт-Петербург", "город-миллионник" и "другие".
    """    
    city = string.split(' , ')[0]
    if city in msk_spb:
        return city
    if city in million_cities:
        return "город-миллионник"
    else:
        return "другие"
    

def reloc_fun(string):
    """Функция, определяющая готовность соискателя к переезду. Задача осложняется тем, что в ряде случаев
    соискатель указывает станцию метро. Ответ определяется по словам "хочу" и "готов" или "не готов"

    Args:
        string (string): значение признака

    Returns:
        bool: ответ на вопрос, готов ли соискатель к переезду 
    """    
    #разбивка аргумента по запятым и выделение информации о готовности к переезду
    #в зависимости от того, указана ли станция метро
    if string.split(' , ')[1].strip()[0:2] == 'м.':
        reloc = string.split(' , ')[2]
    else:
        reloc = string.split(' , ')[1]
    
    return (reloc[0:4] == "гото") or (reloc[0:4] == "хочу")
        
def trip_fun(string):
    """Функция, определяющая готовность соискателя к командировкам. Ее работа опирается на тот факт, что 
    нужная информация -- последняя в строке. Далее она определяет, начинается ли эта подстрока с частицы "не"

    Args:
        string (string): значение признака
        
    Returns:
        bool: ответ на вопрос, готов ли соискатель к командировкам
    """    
    trip = string.split(' , ')[-1].strip() # забираем последнюю подстроку и чистим от пробелов
    return ('командировк' in trip) and (trip[0:5] == 'готов') # на всякий случае проверяем сам факт наличия слова "командировк"


# формируем новые признаки с помощью написанных функций и удаляем старый
hh_database['Город'] = hh_database['Город, переезд, командировки'].apply(city_fun)
hh_database['Готовность к переезду'] = hh_database['Город, переезд, командировки'].apply(reloc_fun)
hh_database['Готовность к командировкам'] = hh_database['Город, переезд, командировки'].apply(trip_fun)
display((hh_database['Готовность к переезду'] * hh_database['Готовность к командировкам']).value_counts(normalize=True))
hh_database = hh_database.drop(columns=['Город, переезд, командировки'], axis=1)
hh_database.info()
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# формируем список типов занятости
employment_list = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']

def empl_fun(string):
    """Функция, создающая массив булевых значений для списка будущих признаков-мигалок занятости

    Args:
        string (string): значение признака

    Returns:
        list of bools: булевы значения для будущих признаков-мигалок занятости
    """    
    bool_list = []
    for i in employment_list:
        bool_list.append(i in string)
    return bool_list

# формируем список типов графика
schedule_list = ['полный день', 'сменный график', 'гибкий график', 'удалённая работа', 'вахтовый метод']

def shed_fun(string):
    """Функция, создающая массив булевых значений для списка будущих признаков-мигалок графика

    Args:
        string (string): значение признака

    Returns:
        list of bools: булевы значения для будущих признаков-мигалок графика
    """    
    bool_list = []
    for i in schedule_list:
        bool_list.append(i in string)
    return bool_list

# пакетно формируем новые признаки с помощью написанных функций и удаляем старые
hh_database['полная занятость'], hh_database['частичная занятость'], hh_database['проектная работа'], hh_database['волонтёрство'], hh_database['стажировка'] = zip(*hh_database['Занятость'].apply(empl_fun))
hh_database['полный день'], hh_database['сменный график'], hh_database['гибкий график'], hh_database['удалённая работа'], hh_database['вахтовый метод'] = zip(*hh_database['График'].apply(shed_fun))
hh_database = hh_database.drop(columns=['Занятость', 'График'], axis=1)

(hh_database['проектная работа'] * hh_database['волонтёрство']).value_counts()

```

В чем наша беда? В том, что помимо желаемой заработной платы соискатель указывает валюту, в которой он бы хотел ее получать, например:

Нам бы хотелось видеть заработную плату в единой валюте, например, в рублях. Возникает вопрос, а где взять курс валют по отношению к рублю?

На самом деле язык Python имеет в арсенале огромное количество возможностей получения данной информации, от обращения к API Центробанка, до использования специальных библиотек, например pycbrf. Однако, это не тема нашего проекта.

* "date" - дата, 
* "proportion" - пропорция, 
* "close" - цена закрытия (последний зафиксированный курс валюты на указанный день).


Перед вами таблица соответствия наименований иностранных валют в наших данных и их общепринятых сокращений, которые представлены в нашем файле с курсами валют. Пропорция - это число, за сколько единиц валюты указан курс в таблице с курсами. Например, для казахстанского тенге курс на 20.08.2019 составляет 17.197 руб. за 100 тенге, тогда итоговый курс равен - 17.197 / 100 = 0.17197 руб за 1 тенге.

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#открываем данные с mdf.ru 
cur_db = pd.read_csv('data/ExchangeRates.csv', sep=',')
#преобразуем его дату в формат datetime
cur_db["date"] = pd.to_datetime(cur_db["date"], dayfirst=True)
#и избавляемся от лишних признаков
cur_db = cur_db.drop(columns=['per', 'time', 'vol'], axis=1)
#cur_db.info()

#переводим "Обновление резюме" в формат datetime
hh_database["Обновление резюме"] = pd.to_datetime(hh_database["Обновление резюме"], dayfirst=True).dt.normalize()


def salary_split(string):
    """Функция разбивает признак 'ЗП' на численное значение зарплаты и наименование валюты.
    Числовое значение пока остается строкой

    Args:
        string (string): значение признака

    Returns:
        list of strings: численное значение зарплаты в виде строки и наименование валюты, указанной соискателем
    """    
    return string.split(' ')

def cur_rename(string):
    """Функция переименовывает валюту согласно ISO-кодировке, используемой в данных с mdf.ru

    Args:
        string (string): значение признака

    Returns:
        string: валюта согласно ISO-кодировке
    """    
    if string == 'бел.руб.':
        return "BYN"
    elif string == 'руб.':
        return "RUB"
    elif string == 'грн.':
        return 'UAH'
    elif string == 'сум':
        return 'UZS'
    else:
        return string
    

# формируем новые признаки с помощью написанных функций 
hh_database['ЗП(число)'], hh_database['ЗП(валюта)'] = zip(*hh_database['ЗП'].apply(salary_split))
hh_database['ЗП(валюта)'] = hh_database['ЗП(валюта)'].apply(cur_rename)
#list(set(hh_database['ЗП(валюта)'])) # проверка списка уникальных валют

#подготавливаем вспомогательную таблицу к объединению и объединяем с основной
cur_db.rename(columns = {'currency':'ЗП(валюта)', 'date':'Обновление резюме'}, inplace = True )
hh_database = hh_database.merge(
    cur_db,
    on=['ЗП(валюта)', 'Обновление резюме'],
    how='left'
)

# добавляем в рублевые признаки единичные коэффициенты для грядущей формулы
hh_database.loc[hh_database['ЗП(валюта)'] == 'RUB', 'close'] = 1
hh_database.loc[hh_database['ЗП(валюта)'] == 'RUB', 'proportion'] = 1
# преобразуем строку с численным значением зарплаты в целые числа
hh_database['ЗП(число)'] = hh_database['ЗП(число)'].astype(int)
# переводим все зарплаты в рубли
hh_database['ЗП (руб)'] = hh_database['ЗП(число)'] * hh_database['close'] / hh_database['proportion']
# удаляем ненужные признаки
hh_database = hh_database.drop(columns=['ЗП', 'proportion', 'close', 'ЗП(валюта)', 'ЗП(число)'], axis=1)
hh_database.info()
```

Поскольку среднее значение находится немного правее моды, распределение с правосторонней асимметрией

```python
# будем искать выбросы методом z-отклонения, с помощью функции outliers_z_score_mod, описанной в учебных модулях
def outliers_z_score_mod(data, feature, log_scale=False, left=3, right=3):
    """Функция поиска выброса методом z-отклонения с возможностью модификации интервала и логарифмирования выборки

    Args:
        data (DaraFrame): таблица данных
        feature (string): признак
        log_scale (bool, optional): Включать ли логарифмирование. Defaults to False.
        left (int, optional): диапазон сигм в левую сторону. Defaults to 3.
        right (int, optional): диапазон сигм в правую сторону. Defaults to 3.

    Returns:
        DaraFrame: таблица со строками-выбросами и очищенная таблица
    """    
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned


outliers, cleaned = outliers_z_score_mod(hh_database, 'Возраст', log_scale=True, left=3, right=4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
hh_database = cleaned
print('Таблица с выбросами')
outliers
```

Метод отсеял двух подростков и одного 100-летнего фронтэндера. На мой взгляд, подростки не являются аномалиями: их данные кажутся реалистичными.
